# Conjuntos de Julia

Los conjuntos de Julia, así llamados por el matemático Gaston Julia, son una familia de conjuntos fractales que se obtienen al estudiar el comportamiento de los números complejos al ser iterados por una función.

Veremos a continuación diferentes ejemplos de las imagenes que se pueden obtener a través de esta técnica.

```python
import julia as ju
```

En su versión más sencilla podemos observar como la función $f(z)=z^2$, nos deja ver como en el intervalo de $[0,1]$ la función converge, mientras que en los valores $z>1$ la función crecerá hasta ir al infinito.

```python
ju.fractal(fn="z**2")
```

Ahora veamos que pasa cuando a la función $f(z)=z^2$ le sumamos un número complejo así $f(z) = z^2 + complex(a,b)$ donde $a, b \in \mathbb{R}$ siendo $a$ la parte real y $b$ la parte imaginaria.

```python
ju.fractal(fn="z**2 + complex(0.2,0.3)")
```

Veamos ahora algunos ejemplos que generan imagenes que han sido reconocidas por su gran belleza. La función es $f(z)=\frac{z^2+z}{ln(z)}+complex(a,b)$

```python
ju.fractal(fn="(z**2+z)/N.log(z) +complex(0.268,0.060)",xa=-3,xb=1,ya=-2,yb=2)
```

Ahora veamos el de la función $f(z) = e^z - 0.65$.

```python
ju.fractal(fn="N.exp(z)-0.65")
```
